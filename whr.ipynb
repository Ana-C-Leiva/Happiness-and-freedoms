```python
df_countries = pd.read_csv('Countries by continents.csv')
```

```python
# Cambiando df_final para coincidir con tabla Countries by continents
    # sacando *
df_final['country'] = df_final['country'].str.replace('*','',regex=False)
    # Cambiando nombres
df_final['country'] = df_final['country'].str.replace('Burkina Faso','Burkina',regex=False)
df_final['country'] = df_final['country'].str.replace('Turkiye','Turkey',regex=False)
df_final['country'] = df_final['country'].str.replace('Congo (Kinshasa)','Democratic Republic of Congo',regex=False)
df_final['country'] = df_final['country'].str.replace('Congo (Brazzaville)' ,'Congo',regex=False)

# Compara las dos columnas de pais - identificar cuales nombres de paises no coinciden
merged_outer = pd.merge(df_final, df_countries, left_on='country', right_on='Country', how='outer', indicator=True)
not_merged = merged_outer[merged_outer['_merge']!= 'both']
valores_unicos_country = not_merged['country'].unique()
valores_unicos_Country = not_merged['Country'].unique()

print(valores_unicos_Country)
print(valores_unicos_country)

df_final.to_csv('whr2015_2024.csv', index=False)
```
